(a) Considera el método de Newton para funciones de números *complejos* $z$. 
Impleméntalo para funciones $f: \mathbb{C} \to \mathbb{C}$.
Considera la función particular $f: z \mapsto z^3 -1$.  Nota que para algunos valores de $z$, el método de Newton *no converge*; por lo tanto, debes asegúrate de  salir de la función después de un número máximo de iterados.**


Como ya se había obtenido con anterioridad, el método de Newton utiliza las  tangentes a la función en puntos cercanos a la raíz y el $n-ésimo$ punto obtenido está determinado por:

$$x_n= x_{n-1}\  -\ \frac{f(x_{n-1})}{f'(x_{n-1})}$$

Si $f'(x_{n-1})$ no es cero. Entonces la función `newton` estpa dada por:

Para utilizar el método de Newton es necesario utilizar la función derivada y en particular para una función compleja $f: \mathbb{C} \to \mathbb{C}$, se utliza la derivada compleja, definida como:

$$ \lim_{z \rightarrow z_0} \frac{f(z)-f(z_0)}{z-z_0} = \lim_{h \rightarrow 0} \frac{f(z+h)-f(z)}{h}$$

Por tanto, defino la función derivada como sigue:



```julia
function derivada(z)
    h=0.000000001
    (f(z+h)-f(z))/h
end
    
```

Dada la restricción de la derivada distinta de cero, la función `newton`, que aplica el método de Newton a las mismas, se  definirá con base en un contador, el cual, se detiene después de un número específico de repeticiones para evitar que ocurra un error cuando la `derivada(z)` es cercana o igual a 0.

```julia
function newton(f,z0,i=0)
    
    z_n= z0-(f(z0)/derivada(z0))  
    i=i+1 
    
    if i<=150
        
        if abs(f(z0))<= 0.0000000001
        z0
        else
        
        newton(f,z_n,i)
        end
    else
        return 0
    end
end    
```

**Método de Euler**

Para resolver esta serie de problemas, primero definiremos una función `euler` que aplique el método de Euler a la función que se le introduzca. En este método cada nuevo término se obtiene a partir de una aproximación de las tangentes a través de la derivada, con base en cierto valor inicial conocido:

$$\xi_{n+1} = \xi_n + hf(t_n,\xi_n)$$

En general, éste método es independiente del número de variables que tenga mi función, pues $\xi$ podría ser un vector $\mathbf{\xi} = (\xi_1,\xi_2,\ldots ,\xi_k)$.

Considerando esto, entonces se tendría:

$$ \begin{pmatrix} \xi_{1(n+1)} \\ \xi_{2(n+1)} \\ \ldots \\ \xi_{k(n+1)}\end{pmatrix} =  \begin{pmatrix} \xi_{1(n)} \\ \xi_{2(n)} \\ \ldots \\ \xi_{k(n)}\end{pmatrix} + h\mathbf f(t_n,\begin{pmatrix} \xi_{1(n)} \\ \xi_{2(n)} \\ \ldots \\ \xi_{k(n)}\end{pmatrix}) $$

Donde $\mathbf f$ es una función determinada por las derivadas.Así pues, es necesaria la construcción de una función que tome vectores como valores para evaluar, lo cual logramos con el comando `vect` haciendo arreglos de arreglos.


```julia
function euler(q0,t0,tf,h,f)
    q = Vector{Float64}[[q0]]
    t = Float64[t0]
    for i in t0:h:tf
        
        q_n = q0 + h*f(t0, q0)
        
        q0 = q_n
        
        push!(t,t0)
        push!(q,[q0])
    end
    t,hcat(q...)'
end
```

**[2] Considera el tiro parabólico. **

**(a) Resuelve las ecuaciones numéricamente y compara con la solución analítica.**



El tiro parabólico, puede considerarse una función vectorial $\mathbf r(t)$  que cumple:

$$F=m\mathbf{a}= m\mathbf{g}=m\ddot{ \mathbf{r}} =m\dot {\mathbf{v}}$$

En donde se tiene que 
$$\mathbf{a} := \begin{pmatrix} 0 \\ -g \end{pmatrix}$$

y de donde pueden obtenerse dos ecuaciones separadas:

 $$\ddot x(t)=0$$ $$\ddot y(t)=\dot v(t)=-g$$ 
 
 Creando los siguientes sistemas para cada ecuación se puede aplicar el método de Euler con los siguientes vectores:
 
$$\mathbf{v} := \begin{pmatrix} v_{0x} \\ v_{0y}-gt \end{pmatrix}$$
 

 

Analíticamente, a través de integración puede llegarse al resultado para $\mathbf r( t)=x(t)\mathbf i+ y(t) \mathbf j $ .Donde se tiene que:

$$x(t)=x_0+v_{0x}t$$

$$y(t)=y_0+v_{0y}t- \frac{1}{2} gt^2$$

Se establece una convención para llamar a los elementos de la matriz **q**:

`q[1]=x0`, posición en x

`q[2]=y0`, posición en y

`q[3]=vx`, velocidad en x

`q[4]=vy`, velocidad en y

Ahora definimos la función vectorial tiro parabólico, cuyas entradas serán $(v_x,v_y,a_x,a_y)$, con a la aceleración.

```julia
function tiro_parabolico(t,q)
    
    g=9.81
    
    [q[3], q[4]-g*t, 0, - g]
    
end
```

```julia
function posicion_analitica(x0,y0,v0x,v0y,t0,tf,h)
  
  
    t=[t0:h:tf]
    
    x=[x0+v0x*t]
    y=[y0+v0y*t-(9.81*t^2)/2 for t in t]    
    
    t,x,y
    
    plot(x,y,"r--",label="Posición analítica",linewidth=3)
    plot(M[:,1], M[:,2],label="Posición aproximada con Euler")
     
    title("Comparación de la solución numérica y la analítica")
    xlabel("Posición en x")
    ylabel("Posición en y")
    
    legend()
        
end
```

**(b) Ahora agrega fricción proporcional a la velocidad. Escribe las EDOs y resuélvelas. Compara con el resultado de (a).**

Si ahora consideramos rozamiento del aire, se tiene que las ecuaciones diferenciales a resolver son:

$$\ddot {v_x}=-\gamma v_{x}$$

$$\ddot {v_y}=-g-\gamma v_{y}$$

Cuyas soluciones analíticas correspondientes son:

$$x(t)= x_0+\frac{v_{x0}}{\gamma} (1-e^{-\gamma t})$$ 

$$y(t)=y_0+\frac{1}{\gamma}(\frac{g}{\gamma}+v_{y0})(1-e^{-\gamma t})-\frac{g}{\gamma}t$$

Manteniendo la misma convención para el orden de las **q** definimos la función tiro_con_rozamiento como sigue:

```julia
function tiro_con_rozamiento(t,q)
    γ=0.2
    g=9.81
    [q[3], q[4],  -γ*q[3], -g-γ*q[4]]
end
```

```julia
condiciones_iniciales = [0, 0, 30*cos(pi/3), 30*sin(pi/3)]
t, M =euler(condiciones_iniciales, 0, 15 , 0.01, tiro_con_rozamiento);
```

```julia
plot(M[:,1], M[:,2])
```

Ahora definimos la función rozamiento analítico aplicando directamente la solución que describe la posición en este caso y comparamos los resultados obtenidos.

```julia
function rozamiento_analitico(x0,y0,v0x,v0y,t0,tf,γ,h)
    c=v0y/γ+(9.81/(γ^2))
    t=[t0:h:tf]
    
    x=[x0+(v0x/γ)*(1-exp(-γ*t)) for t in t]
    y=[y0+c*(1-exp(-γ*t))-((9.8/γ)*t) for t in t]
    
    t,x,y
    
    plot(x,y,"r--",label="Posición con rzamiento analítica",linewidth=3)
    plot(M[:,1], M[:,2],label="Posición con rozamiento aproximada con Euler")
     
    title("Comparación de la solución numérica y la analítica")
    xlabel("Posición en x")
    ylabel("Posición en y")
    
    legend(loc="best") 
    
   
end
```

```julia
rozamiento_analitico(0, 0, 30*cos(pi/3), 30*sin(pi/3),0,15,0.2,0.01)
```

**(c) ¿Cómo cambia si la fricción es proporcional al cuadrado de la rapidez?**



Si es proporcional al cuadradoo de la rapidez, las ecuaciones están dadas por:
 $$\ddot y = -g-γ\dot y \sqrt{\dot x^2 + \dot y^2}$$
$$\ddot x = -γ\dot x \sqrt{\dot x^2 + \dot y^2}$$

Donde la expresión $\sqrt{\dot x^2 + \dot y^2} = v$ es la rapidez 

```julia
function tiro_con_rozamiento2(t,q)
    γ=0.2
    g=9.81
    
    [q[3], q[4], -γ*q[3]*sqrt(q[3]^2+ q[4]^2),  -g-γ*q[4]*sqrt(q[3]^2+ q[4]^2)]
end    
```

```julia
condiciones_iniciales = [0, 0, 30*cos(pi/3), 30*sin(pi/3)]
t, M =euler(condiciones_iniciales, 0, 15 , 0.01, tiro_con_rozamiento2);
```

```julia
plot(M[:,1], M[:,2]) 
title("Tiro con rozamiento proporcional al cuadrado de la rapidez")
 xlabel("Posición en x")
ylabel("Posición en y")
legend()
```

```julia
condiciones_iniciales = [0, 0, 30*cos(pi/3), 30*sin(pi/3)]
t, M1 =euler(condiciones_iniciales, 0, 15 , 0.01, tiro_con_rozamiento2);

t, M2 =euler(condiciones_iniciales, 0, 15 , 0.01, tiro_con_rozamiento);

t, M3 =euler(condiciones_iniciales, 0, 15 , 0.01, tiro_parabolico);

plot(M3[:,1], M3[:,2],label="Tiro parabólico")
plot(M2[:,1], M2[:,2],label="Tiro con rozamiento")
plot(M1[:,1], M1[:,2], label= "Rozamiento proporcional al cuadrado de v")
legend()
```
