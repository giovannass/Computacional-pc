**[1] Escribe una función recursiva para calcular una raiz de una función $f$ en un intervalo $[a,b]$ con el método de bisección. Verifica que funcione al calcular raices de distintas funciones.**

**RESPUESTA**

El método de bisección requiere conocer un intervalo por el cual se sabe que la función cambia de signo, es decir, que cruza el cero. Además es recomendable que sea sólo una raíz la que hay en el intervalo, pues  sólo puede calcularse una de ellas y podría haber errores al ejecutar la función. La función propuesta `raiz_bis(f,a,b)` requiere que se defina primero la función $f(x)$ para posteriormente asignar los extremos $a$ y $b$ del intervalo y comenzar la bisección para encontrar el valor de $x$ donde $f$ es igual a cero.

```julia
function raiz_bis(f,a,b)
    
if f(a)*f(b)>0

    println("Para que el método funcione, la función evaluada en los extremos del intervalo debe tener signos distintos")

elseif f(a)*f(b)<0

        x=(1/2)*(a+b)

        if abs(f(x))<=0.000000001
            
            x

        elseif f(a)*f(x)<0
            
            a1=a
            b1=x
            
            raiz_bis(f,a1,b1)
            
        elseif f(x)*f(b)<0
            
            a1=x
            b1=b
            
            raiz_bis(f,a1,b1)
         end   
         
else

    e1 = f(a)
    e2 = f(b)

        if e1 == 0

            println("El extremo a del intervalo es la raíz buscada")           
        
        elseif e2== 0
    
            println("El extremo b del intervalo es la raíz buscada")
        end
    end
end

            
        
        
```

**[2] Considera la colisión en una dimensión de una partícula haciendo movimiento armónico simple con frecuencia angular $\omega$ y una partícula libre con velocidad $v$.**

(a) Escribe una función que toma los parámetros del oscilador y de la partícula y grafica sus respectivas posiciones en el tiempo, suponiendo que no hay colisión. 

(b) Dibuja (en gráficas separadas) las trayectorias de las dos partículas para unas cuantas posibilidades de los parámetros. ¿Cuántas "colisiones" puede haber? ¿Cuál es la buena?

(c) Escribe una función para calcular el tiempo de colisión de las dos partículas, sin garantizar cuál es, utilizando la función que escribiste en la pregunta [1].


**RESPUESTA**

a) Un oscilador armónico es un sistema donde una masa se encuentra sometida a una fuerza restitutiva

$$F=-kx$$

que conduce a la ecuación diferencial 

$$ \ddot{x}+\omega^2 x=0$$ 
Donde $\omega\ = \ \sqrt{\frac{k}{m}}$ 

La solución a dicha ecuación está dada por:

$$ x(t)\ = \ A\ \cos(\omega t)\ + \ B\ \sin(\omega t)$$

Donde $$\ddot{x}(t)\ =\ -\omega A\ \sin(\omega t) + \omega B\ \cos(\omega t)$$


Aplicando condiciones iniciales se encuentran los respectivos valores de $A$ y $B$:

$$x(0)=A=x_0$$

$$\ddot{x}(0)= \omega B\ = v_0$$

Por otra parte una partícula libre se representa por la ecuación:

$$x(t)=x_0\ +\ vt$$
Donde $v$ es la velocidad constante.

Para el caso de la colisión, la partícula debe dirigirse hacia el oscilador, por lo que se tomará una velocidad negativa $-v$, es decir, la ecuación de nuestra partícula estará dada por $$x(t)=x_0\ -\ vt$$

y la colisión se presentará cuando las posiciones de ambas partículas concidan, es decir cuando se tenga 

$$  A\ \cos(\omega t)\ + \ B\ \sin(\omega t) =\ x(t)=x_0\ -\ vt$$

$$  A\ \cos(\omega t)\ + \ B\ \sin(\omega t) -\ x_0\ +\ vt=\ 0$$


El valor de $t$ para el cual se cumpla la última ecuación será el tiempo de colisión $t_c$.


Para resolver la ecuación se aplicará el método de bisección ya mencionado.

**FUNCIÓN PARA EL OSCILADOR**

```julia
function oscilador(ω,x0,v0)
    A=x0
    B=v0/ω
    t=0:0.1:20
    A*cos(ω*t) + B*sin(ω*t)
 end
```

**GRÁFICA CORRESPONDIENTE**

(a) Verifica que tu función sirve para encontrar raices de la función $f(x) = x^2 - 2$.

**RESPUESTA**





El método de Newton utiliza las  tangentes a la función en puntos cercanos a la raíz, las cuales son construidas a partir de un punto inicial $x_1$, el valor de la función evaluada $f(x_1)$ y la pendiente a la tangente en ese punto $f'(x_1)$ . Se intersectan las tangentes al eje corrdernado y en el punto de interseción se comienza el proceso de nuevo. El $n-ésimo$ punto obtenido está determinado por:

$$x_n= x_{n-1}\  -\ \frac{f(x_{n-1})}{f'(x_{n-1})}$$

Si $f'(x_{n-1})$ no es cero. De esta forma contruimos la función `newton` como sigue:

```julia
function newton(f,x0)
    
    if abs(f(x0))<= 0.00000001
        x0
    else
        
    x_n= x0-(f(x0)/df(x0))
        newton(f,x_n)
    end
end
```


Donde x0 es la adivinanza, f la función previamente definida y df la derivada de la misma

```julia
f(x)=x^2-2
```

```julia
df(x)=2*x
```

```julia
newton(f,4)
```

```julia
newton(f,55)
```

```julia
newton(f,20)
```

```julia
newton(f,-3)
```

```julia
newton(f,-16)
```

```julia
newton(f,-48)
```

Estos resultados muestran que llega a la raíz mas cercana a la adivinanza, además de que si funciona correctamente para varios valores iniciales.

```julia
function tangente(f,x0,x)

    (df(x0))*(x-x0)+f(x0)
        
end
```

```julia
X=[-10:0.2:10]
```

```julia

figure(figsize=(5,4))

X=[-15:0.2:15]
X0=[4:5:20]
Y =[tangente(f,x0,x) for x in X, x0 in X0]
funcion=[f(x) for x in X]
plot(X,Y, label="Tangente a x0")
plot(X,funcion, label= "función f(x)")

legend(loc=0, fontsize="x-small")
title(L"$Aproximaciones\ \ con\ \ x0>0$")
xlabel(L"$x$")
ylabel(L"$f(x)$")

ylim(-300, 400)
```

```julia

figure(figsize=(5,4))

X=[-15:0.2:15]
X0=[-20:5:3]
Y =[tangente(f,x0,x) for x in X, x0 in X0]
funcion=[f(x) for x in X]
plot(X,Y, label="Tangente a x0")
plot(X,funcion, label= "función f(x)")

legend(loc=0, fontsize="x-small")
title(L"$Aproximaciones\ \ con\ \ x0<0$")
xlabel(L"$x$")
ylabel(L"$f(x)$")

ylim(-300, 400)
```

En las gráficas se observa como las tangentes al depender de la inclinación de la pendiente, se acercan más a la raíz positiva, o a la negativa, dependiendo del valor inicial.
