##  Metodo de Euler

Empieza en $x(t=0)=x_0 \ en\ t_0$
Idea el metodo:

Dar un paso de tamaño "h" en el tiempo, i.e. $x(t_o+h)$

$x_1=x_0+hf(x,t_0)$

$x_2=x_1+h +hf(x_1,t_1)$ con $t_1=t_0+h$

Vamos a resolver $\dot x = -x= f(x,t)$

```julia

f(x,t)=-x
```

```julia
t0=0.0
x0=1.0
t_final= 5.
h=0.1

num_pasos= int((t_final-t0)/h)
 #

ts=[t0:h:t_final]
num_pasos= length(ts)
xs=zeros(ts)

for n in 1:num_pasos
   
    xs[n]=x_actual
    
    x_actual=x0
    t_actual=t0

    x_nueva=x_actual+h*f(x_actual,t_actual)
    t_nueva=t_actual+h

    @show (x_nueva,t_nueva)
    x_actual=x_nueva
    t_actual=t_nueva
end
```
