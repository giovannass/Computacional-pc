```julia
#CONVENCION:
#Q[1]=x
#Q[2]=vx
#Q[3]=y
#Q[4]=vy
function tiro_parabolico(t,Q)
    λ=0.1
    g=9.8
    [Q[2], -λ*Q[2]*sqrt(Q[2]^2+ Q[4]^2), Q[4], -g-λ*Q[4]*sqrt(Q[2]^2+ Q[4]^2)]
end
```

```julia
function euler(y0,t0,h,f,N)
    y = Vector{Float64}[[y0]]
    t = Float64[t0]
    for i in 1:N
        y_nueva = y0 + h*f(t0, y0)
        t0 = t0 + h
        y0 = y_nueva
        push!(t,t0)
        push!(y,[y0])
    end
    t,hcat(y...)'
end
```

```julia
c_ini = [0, 30*cos(pi/3), 0, 30*sin(pi/3)]
t, M =euler(c_ini, 0, 0.001 ,tiro_parabolico, 15000);
```

```julia
function euler_cromer(q0,v0,t0,tf,h,f_q,f_v)
    q = Vector{Float64}[[q0]]
    v = Vector{Float64}[[v0]]
    t = Float64[t0]
    for i in t0:h:tf
        v_n= v0 + h*f_v(t0, q0, v0)
        q_n = q0 + h*f_q(t0, q0,v_n)
        
        
        q0 = q_n
        v0= v_n
        
        push!(t,t0)
        push!(q,[q0])
        push!(v,[v0])
        
        if abs (q0[1])> pi
            break
        end
    end
    t,hcat(q...)', hcat(v...)'
end
```

```julia
#CONVENCION:
#q[1]=x
#q[2]=y
#v[1]=vx
#v[2]=vy
function f_q(t,q,v)
    
    [v[1],v[2]]
    
end

function f_v(t,q,v)
    λ=0.1
    g=9.8
    [ -λ*v[1]*sqrt(v[2]^2+ v[2]^2)  , -g-λ*v[2]*sqrt(v[1]^2+ v[2]^2)]
    
end
```

```julia
q_ini = [0, 0]
    
v_ini =  [30*cos(pi/3),  30*sin(pi/3)]
t, q, v =euler_cromer(q_ini, v_ini, 0, 10 ,0.001, f_q,f_v);
```

```julia
plot(q[:,1],q[:,2])
```

```julia
function f_q(t,q,v)
  
    v 
    
end

function f_v(t,q,v)
    ℓ=13.5
    (-9.8/ℓ)*sin(q)
end

t,v,q =euler_cromer(pi-0.01, 0, 0, 100 ,0.001, f_q,f_v);
```
